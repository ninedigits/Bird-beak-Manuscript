```python
import pandas as pd
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams.update({'font.size': 25})
import seaborn as sns
plt.style.use('ggplot')
%matplotlib inline
```

```python
# Read in zones and bird-beak data.
zones = pd.read_csv('zones_group_bbh.csv')
zones = zones.set_index('pat_num')
zones
```

```python
#zones['bbh'].hist()
sns.distplot(zones['bbh'], bins=20)
```

```python
fig, (ax1, ax2) = plt.subplots(2,1, figsize=(10, 10))
colors = color=sns.color_palette('cividis',4)
scatter_kws = colors[0]
line_kws = colors[-1]
sns.regplot(x='CD', y='bbh', data=zones, ax=ax1, scatter_kws={"color": scatter_kws}, line_kws={"color": line_kws})
ax1.set_xticklabels([])
ax1.set_xlabel('')

ax1.set_ylabel('Bird-Beak Height (mm)', fontsize=20)
sns.regplot(x='CD', y='bba', data=zones, ax=ax2, scatter_kws={"color": scatter_kws}, line_kws={"color": line_kws})
ax2.set_xlabel('Curvature × Diameter', fontsize=20)
ax2.set_ylabel('Bird-Beak Angle (deg)', fontsize=20)
plt.savefig('/Users/maxfrohlich/Dropbox/Manuscripts/Bird-beak Paper/Presentations/corr.png', dpi=400, bbox_inches='tight')
```
